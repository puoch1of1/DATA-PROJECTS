```python
import matplotlib.pyplot as plt
from pathlib import Path
import sys

# Ensure import path includes Data Migration folder (relative to notebooks/)
proj_root = Path.cwd().parent
data_migration_dir = proj_root / "Data Migration"
sys.path.append(str(data_migration_dir))

from validate import get_sqlserver_count, get_postgres_count

try:
    sqlserver_count = get_sqlserver_count("customers", schema="dbo")
    postgres_count = get_postgres_count("customers")
except Exception as e:
    print(f"Count retrieval failed: {e}")
    sqlserver_count = 0
    postgres_count = 0

databases = ["SQL Server", "PostgreSQL"]
counts = [sqlserver_count, postgres_count]

plt.figure()
plt.bar(databases, counts, color=["#1f77b4", "#2ca02c"])
plt.title("Row Count Comparison After Migration")
plt.ylabel("Number of Rows")
plt.xlabel("Database")
for i, c in enumerate(counts):
    plt.text(i, c, str(c), ha="center", va="bottom")
plt.show()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/south_sudan.db")

query = """
SELECT year, value
FROM socio_economic_indicators
WHERE indicator = 'Population'
ORDER BY year
"""

df = pd.read_sql(query, conn)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure()
plt.plot(df["year"], df["value"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("South Sudan Population Over Time")
plt.show()
```

```python
query = """
SELECT year, value
FROM socio_economic_indicators
WHERE indicator = 'GDP (current US$)'
ORDER BY year
"""
gdp = pd.read_sql(query, conn)
```

```python
plt.figure()
plt.plot(gdp["year"], gdp["value"])
plt.xlabel("Year")
plt.ylabel("GDP (current US$)")
plt.title("South Sudan GDP Over Time")
plt.show()
```

```python
query = """
SELECT year, value
FROM socio_economic_indicators
WHERE indicator = 'Life Expectancy at Birth'
ORDER BY year
"""
life_exp = pd.read_sql(query, conn)
```

```python
print(life_exp.head())
print(f"\nShape: {life_exp.shape}")
print(f"\nData types:\n{life_exp.dtypes}")
print(f"\nNull values:\n{life_exp.isnull().sum()}")
```

```python
# Check what indicators exist in the database
query_check = """
SELECT DISTINCT indicator
FROM socio_economic_indicators
"""
indicators = pd.read_sql(query_check, conn)
print(indicators)
```
